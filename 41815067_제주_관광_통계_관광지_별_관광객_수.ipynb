```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

combined_df['userCnt'] = pd.to_numeric(combined_df['userCnt'], downcast='float')

fig = plt.figure(figsize=(14, 6))
ax = plt.subplot(1, 1, 1)

ax.plot(df1['dtYearMonth'], df1['userCnt'], color='red', lw=1.5, alpha=0.7, label='201501-201512')
ax.plot(df2['dtYearMonth'], df2['userCnt'], color='blue', lw=1.5, alpha=0.7, label='201601-201612')
ax.plot(df3['dtYearMonth'], df3['userCnt'], color='yellow', lw=1.5, alpha=0.7, label='201701-201712')
ax.plot(df4['dtYearMonth'], df4['userCnt'], color='orange', lw=1.5, alpha=0.7, label='201801-201812')
ax.plot(df5['dtYearMonth'], df5['userCnt'], color='green', lw=1.5, alpha=0.7, label='201901-201912')
ax.plot(df6['dtYearMonth'], df6['userCnt'], color='black', lw=1.5, alpha=0.7, label='202001-202012')
ax.plot(df7['dtYearMonth'], df7['userCnt'], color='grey', lw=1.5, alpha=0.7, label='202101-202112')

ax.set_xlabel('연도 별', fontweight='bold', fontsize=12)
ax.set_ylabel('관광지별 외국인 유입자 수', fontweight='bold', fontsize=12)
ax.set_title('연도별 관광지 외국인 유입자 비교', fontweight='bold', fontsize=12)

# x축 라벨 설정
combined_dates = pd.concat([df1['dtYearMonth'], df2['dtYearMonth'], df3['dtYearMonth'], df4['dtYearMonth'], df5['dtYearMonth'],df6['dtYearMonth'],df7['dtYearMonth']])
combined_dates = pd.to_numeric(combined_dates, errors='coerce')
x_labels = combined_dates.dropna().unique()
x_labels = np.sort(x_labels)
ax.set_xticks(np.arange(len(x_labels)))
ax.set_xticklabels(x_labels, rotation=45)

plt.yticks(np.arange(0, 35000, 5000))

# 범례 표시
ax.legend()

plt.show()
```
